## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## ML Task Type

This project is best framed as a **ranking** task. The goal is to rank content pages by their refresh priority so that editors can review the most important pages first. Instead of making only a yes/no decision, the model produces an ordered list that helps prioritize limited editorial resources.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target (Proxy)

The target is whether a content page should be considered a high refresh priority. In the starter dataset, I use **declining content** as a proxy target by using the provided trend information. This is an observed label derived from the dataset rather than a manually defined business rule. It helps identify pages that may need editorial attention first.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

The primary success metric is **Precision@20**. Since the goal is to recommend the highest-priority pages for review, this metric measures how many of the top 20 recommended pages are truly high-priority according to the chosen proxy. A higher Precision@20 means editors spend more time reviewing the right pages and less time on unnecessary work.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A fixed rule, such as refreshing every page older than 180 days or every page with low impressions, cannot capture all situations. Content refresh priority depends on multiple factors working together, including impressions, content age, ranking position, CTR, and update history. A machine learning model can learn how these signals interact and rank pages more effectively than a single if-statement, helping editors focus on the pages that are most likely to benefit from a refresh.
